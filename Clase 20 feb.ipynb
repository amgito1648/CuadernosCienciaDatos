```python
# guarde en una variable el api del dataset de nmist y cargue en la variable ( x_train, y_train),(X_test, y_test) ese dataset
(x_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# muestre la dorma de un numero del arreglo X_train y de un numero del arreglo y_train
print("Forma de X_train:", x_train[0].shape)
print("Forma de y_train:", y_train[0].shape)
```

```python
# crear una red neuronal secuencial con una capa de entrada de 784 neuronas, una capa oculta de 128 neuronas y una capa de salida de 10 neuronas
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Capa de entrada que aplana la imagen de 28x28 a un vector de 784
    tf.keras.layers.Dense(128, activation='relu'),  # Capa oculta con 128 neuronas y función de activación ReLU
    tf.keras.layers.Dense(10, activation='softmax')  # Capa de salida con 10 neuronas y función de activación softmax
])
# compile el modelo con el optimizador adam, la función de pérdida sparse_categorical_crossentropy y la métrica de precisión
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# entrene el modelo con los datos de entrenamiento durante 5 épocas
model.fit(x_train, y_train, epochs=5)
# evalúe el modelo con los datos de prueba
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Pérdida en el conjunto de prueba:", test_loss)
print("Precisión en el conjunto de prueba:", test_acc)

```

```python
model.save('model.keras')
```
